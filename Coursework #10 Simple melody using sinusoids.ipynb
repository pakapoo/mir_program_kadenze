```python
#Assignment-Write a Python function that takes as argument a MIDI note number and duration in seconds and creates a sinusoid signal with the corresponding frequency and duration.
import numpy as np
import IPython.display as ipd
srate = 44100.0

#Generate sinusoid
def sinusoid(freq=440.0, dur=5.0, srate=srate, amp=1.0, phase = 0.0):  
    t = np.linspace(0,dur,int(srate*dur))       #initialize numpy array with number of samples 
    data = amp * np.sin(2*np.pi*freq *t+phase)  #generate sinusoids with given freq, amplitude, phase
    return data 

#Generate single note with MIDI number
def gen_note(midi_num: int, duration: float = 1, srate: float = srate):
    ##############################################################################################
    #When two notes are relationship of a octave: f2/f1 = 2 => number of octaves = log2(f2/f1)   #
    #f2 = 2^(number of octaves)*f1                                                               #
    #Since there are 12 semitones in a octave; plus f1 is usually referred to A4 (440Hz)         #
    #=> f2 = 2^(number of semitones from A4/12)*440                                              #
    #=> number of semitones from A4 = 12*log2(f2/440)                                            #
    #Since A4 is 69 in MIDI number, and MIDI number increased by 1 for a semitone                #
    #=> f2 = 2^(m-69/12)*440                                                                     #
    ##############################################################################################
    frequency = pow(2, (midi_num-69)/12) * 440
    return sinusoid(freq=frequency, dur=duration)

#Generate melody with a list of MIDI number
def gen_melody(midi_num_list: list, duration: float = 1):
    melody = [gen_note(i, duration) for i in midi_num_list]
    return np.hstack(melody) #use hstack for numpy array
```

```python
####################
# play single note #
####################
midi_num = 80
note = gen_note(80, 3)
ipd.Audio(note, rate=srate)
```

```python
###############
# play melody #
###############
midi_seq = [69, 71, 73, 74, 76]
melody = gen_melody(midi_seq)
ipd.Audio(melody, rate=srate)
```
